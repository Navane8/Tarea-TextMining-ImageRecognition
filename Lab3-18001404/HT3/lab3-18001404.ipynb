## Problema 1:

Implemente una regex para validar un correo electronico en general, a continuacion se muestran algunos ejemplos.
- Guate.360-porelmundo@miguate.com
- Miercoles3@hotmail.com
- Progra3.galileo@galileo.edu

```python
def ValidaCorreo(correo):
    if re.match(r'^[\w\.-]+@[\w\.-]+\.\w+$', correo):
        print(f'El correo: {correo} es un correo valido')
    else:
        print(f'El correo: {correo} es un correo no valido')
```

```python
correos=['Guate.360-porelmundo@miguate.com','Miercoles3@hotmail.com','Progra3.galileo@galileo.edu','noval@','no.com','no2@.']

for corr in correos:
    ValidaCorreo(corr)
```

## Problema 2

Implemente una regex para validar la direccion url de una pagina web con los tipos de domino (.com, .org, .edu). Note que la url incluye el protocolo (http o https) y los sımbolos (//www.), a continuacion se muestran algunos ejemplos:
- https://www.guate360-porelmundo.com
- http://www.a2.net 
- https://www.galileo.edu
- http://www.8.org (No valida)

```python
def ValidaPagina(pag):
    if re.match(r'https?://www\.[^0-9][\w-]+\.(com|edu|org)$', pag):
        print(f'La pagina: {pag} es una pagina valida')
    else:
        print(f'La pagina: {pag} es una pagina no valida')
```

```python
paginas=['https://www.3guate360-porelmundo.com',' http://www.a2.net', 'https://www.galileo.edu','http://www.8.org']
for pag in paginas:
    ValidaPagina(pag)
```

## Problema 3

Implemente una regex para validar una MAC Address, notar que las mac address estan
divididas en 6 bloques de caracteres hexadecimales, es decir que los sımbolos solo pueden
variar del 0 al 9 y las letras de la A a la F. a continuacion se muestran algunos ejemplos:
- 5A 6F AF 8C 9B 1D
- 6D 6C 4D 3A EB 3F
- 3A 7C FA C8 6D 4J (no valida por que el ultimo bloque contiene una J)

```python
def ValidaMAC(celda):
    if re.match(r'^((\d[A-F]|[A-F]\d[A-F]|[A-F]{2})\s+){5}\d[A-F]|[A-F]\d[A-F]|[A-F]{2}$', celda):
        print(f'La MAC Address: {celda} es una MAC Address valida')
    else:
        print(f'La MAC Address: {celda} es una MAC Address no valida')
```

```python
celdas=['5A 6F AF 8C 9B 1D','6D 6C 4D 3A EB 3F','3A 7C FA C8 6D 4J']
for c in celdas:
    ValidaMAC(c)
```

## Problema 4
Implemente una regex para validar una direccion IPv4, notar que las direcciones IPv4 estan divididas en 4 bloques de valores los cuales solo pueden ir desde 0 hasta 255, una ip donde algunos de sus bloques sea mayor a 255 no es valida, ademas tome en cuenta que cada bloque esta separada por un punto. A continuacion se muestran algunos ejemplos:
- 192.16.8.1
- 234.56.78.90
- 1.2.3.4
- 192.168.45.345 (no valida por que el ultimo bloque es mayor a 255)


```python
def ValidaIP(ip):
    if re.match(r'^(25[0-5]|2[0-4]\d|1\d\d|[1-9]\d|\d)(\.(25[0-5]|2[0-4]\d|1\d\d|[1-9]\d|\d)){3}$', ip):
        print(f'La IP: {ip} es una IP valida')
    else:
        print(f'La IP: {ip} es una IP no valida')
```

```python
ips=['192.16.8.1','234.56.78.90','1.2.3.4','192.168.45.345']
for ip in ips:
    ValidaIP(ip)
```

# Serie 2
En la carpeta encontrara adjuntos 21 documentos que tiene 100 fechas en la secuencia dıas-mes-año pero con distinto separador y distinto formato de mes, en algunos casos aparece un numero y en otros el nombre del mes en ingles, por ejemplo: Enero pueden aparecer como 1 o como Jan. Utilice Python y expresiones regulares para encontrar el dıa, mes y año promedio total del los 21
archivos, los resultados deben ser un double.

```python
# Ruta de la carpeta
carpeta= 'Datos/'
# dataframe final que resultara de la union de varios dataframes
df_final = pd.DataFrame(columns=['Fechas'])

# Ciclo para leer archivos de texto
for archivo in os.listdir(carpeta):
    if archivo.endswith(".txt"):  
        ruta_archivo = os.path.join(carpeta, archivo)
        with open(ruta_archivo, "r") as file:
            contenido= file.read()
            filas = contenido.split('\n')
            #crea dataframe con el contenido del archivo
            df= pd.DataFrame({'Fechas': filas})
            #como tiene mismo nombre de columna, se concatena para que resulte en un dataframe final con todos los df por archivo
            df_final = pd.concat([df_final, df], ignore_index=True)
```

```python
def ConvFecha(texto):
    # Patrón para detectar diferentes formatos de fecha
    patron= r'^((3[0-1]|2[0-9]|1[0-9]|[1-9]|0?[1-9])[./-]?)(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Set|Oct|Nov|Dec|1[0-2]|[1-9]|0?[1-9])[./-]?(20[0-9]{2}|[1-2][0-9]|0?[1-9])$'
    # Buscar todas las coincidencias del patrón en el texto
    coincidencias = re.findall(patron, texto)
    # Convertir las fechas a un formato común (dd-mm-yyyy)
    fechas_convertidas = []
    
    for coincidencia in coincidencias:
        _, dia, mes, anio = coincidencia
        if mes =='Jan':
            mes='1'
        elif mes=='Feb':
            mes='2'
        elif mes=='Mar':
            mes='3'
        elif mes=='Apr':
            mes='4'
        elif mes=='May':
            mes='5'
        elif mes=='Jun':
            mes='6'
        elif mes=='Jul':
            mes='7'
        elif mes=='Aug':
            mes='8'
        elif mes=='Sep'or mes=='Set':
            mes='9'
        elif mes=='Oct':
            mes='10'
        elif mes=='Nov':
            mes='11'
        elif mes=='Dec':
            mes='12'
        else:
             mes
        fecha = f"{dia}-{mes}-{anio}"
        fechas_convertidas.append(fecha)
    
    return fechas_convertidas
```

```python
fechasTransformadas=[]
for fecha in df_final['Fechas']:
    fechasTransformadas.append(ConvFecha(fecha))
fechasTransformadas
```

```python
#verifico que se transformaron todas las fechas (misma cantidad de filas, mismo len de la lista)
len(fechasTransformadas)
```

```python
#incializo variables
promDia = 0
promMes = 0
promAnio = 0
count = 0

for fechalist in fechasTransformadas:
    if fechalist:
        fecha = fechalist[0]
        fechaconv = datetime.strptime(fecha, "%d-%m-%Y")
        promDia += fechaconv.day
        promMes += fechaconv.month
        promAnio += fechaconv.year
        count += 1

if count > 0:
    promDia /= count
    promMes /= count
    promAnio /= count

print("Promedio de día:", promDia)
print("Promedio de mes:", promMes)
print("Promedio de año:", promAnio)
```
